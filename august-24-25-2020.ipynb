```python
r_left = pd.merge(df1,df2,on='DATE_TIME',how='left')
r_left
```

```python
r_right = pd.merge(df1,df2,on='DATE_TIME',how='right')
r_right
```

```python
r_right.info()
```

```python
r_right['DC_POWER'].isnull().value_counts()
```

```python
r_left['IRRADIATION'] = r_left['IRRADIATION'].fillna(0)
r_left['AMBIENT_TEMPERATURE'] = r_left['AMBIENT_TEMPERATURE'].fillna(0)
r_left['MODULE_TEMPERATURE'] = r_left['MODULE_TEMPERATURE'].fillna(0)

X = r_left.iloc[:, 15:16].values #Irradiation
y = r_left.iloc[:, 3].values #DC_Power
```

```python
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2, random_state=0)
```

```python
X_train.shape
```

```python
X_test.shape
```

```python
X_train.ndim
```

```python
y_train.shape
```

```python
y_test.shape
```

```python
from sklearn.linear_model import LinearRegression
lin_reg = LinearRegression()
lin_reg.fit(X_train, y_train)
```

```python
y_pred = lin_reg.predict(X_test)
y_pred
```

```python
y_test
```

```python
plt.scatter(X_test, y_test,  color='gray')
plt.scatter(X_test, y_pred, color='red', linewidth=2)
plt.show()

```
